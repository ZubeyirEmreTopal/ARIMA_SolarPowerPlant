```python
import pandas as pd
```

```python
df=pd.read_csv('inv_verii.txt')
```

```python
df['time']=pd.to_datetime(df['time'])
```

```python
df['time']
```

```python
for i in range (len(df)):
    if pd.isnull(df['status'][i])==True:
        df['status'][i]=512
        print(df['status'][i])
```

```python
for i in range (len(df)):
    if pd.isnull(df['error'][i])==True:
        df['error'][i]=0
        print(df['error'][i])
        
```

```python

for i in range (len(df)):
    if pd.isnull(df['dc_power'][i])==True:
        df['dc_power'][i]=18.03
        print(df['dc_power'][i])
```

```python

for i in range (len(df)):
    if pd.isnull(df['eac'][i])==True:
        df['eac'][i]=1.33
        print(df['eac'][i])
```

```python
df=df.drop(['device_id'],axis=1)
```

```python
el_df=df.set_index('time')
```

```python
final_df=el_df.resample('H').last()
```

```python
final_df
```

```python
import pmdarima as pm
```

```python
final_df['e_day']
```

```python
model = pm.auto_arima(final_df['e_day'], 
                        m=12, seasonal=True,
                      start_p=0, start_q=0, max_order=4, test='adf',error_action='ignore',  
                          suppress_warnings=True,
                      stepwise=True, trace=True)
```

```python
model
```

```python
df['time'][100000]
```

```python
train=final_df[(final_df.index.get_level_values(0) >= '2021-04-25 3:00:00') & (final_df.index.get_level_values(0) <= '2022-04-07 08:15:00')]
```

```python
len(train)
```

```python
test=final_df[(final_df.index.get_level_values(0) > '2022-04-07 08:15:00')]
```

```python
len(test)
```

```python
from statsmodels.tsa.arima_model import ARIMA
model=ARIMA(train['e_day'],order=(5,0,1),seasonal_order=(2,0,2,12),supress_warning=True)
model=model.fit()
model.summary()
```

```python
forecast=model.predict(n_periods=1895, return_conf_int=True)
```

```python
forecast[0]
```

```python
forecast_df = pd.DataFrame(forecast[0],columns=['Prediction'],index=test.index)
```

```python
test.index
```

```python
forecast_df
```
